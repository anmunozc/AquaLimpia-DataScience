**FLUJO DE TRABAJO REPRODUCIBLE**

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def crear_flujo_trabajo():

    #========================
    # CONFIGURACIÓN GENERAL
    #========================

    fig, ax = plt.subplots(figsize=(10, 12), dpi=300)

    ax.set_xlim(0, 10)
    ax.set_ylim(0, 12)

    ax.axis('off')

    # Fondo general
    fig.patch.set_facecolor('#F4F6F7')

    #========================
    # PALETA DE COLORES
    #========================

    COLOR_OBJETIVO = '#FF6B35'
    COLOR_CONFIG = '#1F3C88'
    COLOR_PROCESO = '#3A86FF'
    COLOR_OPERACIONES = '#06D6A0'
    COLOR_AMBIENTAL = '#2ECC71'
    COLOR_DASHBOARD = '#9B5DE5'
    COLOR_GITHUB = '#222222'

    COLOR_TEXTO = 'white'
    COLOR_FLECHA = '#2C3E50'

    #========================
    # FUNCIÓN PARA CREAR CAJAS
    #========================

    def crear_caja(x, y, ancho, alto, titulo, subtitulo, color):

        # -------------------------
        # SOMBRA
        # -------------------------

        sombra = patches.FancyBboxPatch(
            (x + 0.06, y - 0.06),
            ancho,
            alto,
            boxstyle="round,pad=0.12",
            facecolor='gray',
            alpha=0.25,
            linewidth=0
        )

        ax.add_patch(sombra)

        # -------------------------
        # CAJA PRINCIPAL
        # -------------------------

        caja = patches.FancyBboxPatch(
            (x, y),
            ancho,
            alto,
            boxstyle="round,pad=0.12",
            facecolor=color,
            edgecolor='#34495E',
            linewidth=1.5
        )

        ax.add_patch(caja)

        # -------------------------
        # TÍTULO
        # -------------------------

        ax.text(
            x + ancho/2,
            y + alto*0.63,
            titulo,
            ha='center',
            va='center',
            fontsize=11,
            fontweight='bold',
            color=COLOR_TEXTO
        )

        # -------------------------
        # SUBTÍTULO
        # -------------------------

        ax.text(
            x + ancho/2,
            y + alto*0.28,
            subtitulo,
            ha='center',
            va='center',
            fontsize=9,
            color=COLOR_TEXTO
        )

    #========================
    # FUNCIÓN PARA CREAR FLECHAS
    #========================

    def crear_flecha(x1, y1, x2, y2):

        ax.annotate(
            '',
            xy=(x2, y2),
            xytext=(x1, y1),
            arrowprops=dict(
                arrowstyle='->',
                lw=3,
                color=COLOR_FLECHA
            )
        )

    #========================
    # TÍTULO PRINCIPAL
    #========================

    plt.title(
        "Flujo de Trabajo Reproducible - AquaLimpia S.A.",
        fontsize=16,
        fontweight='bold',
        color='#1B2631',
        pad=30
    )

    #========================
    # BLOQUES DEL FLUJO
    #========================

    # -----------------------------------
    # OBJETIVO ANALÍTICO
    # -----------------------------------

    crear_caja(
        3.0, 10.5, 4.0, 0.9,
        "Objetivo Analítico",
        "Diagnóstico operacional y ambiental",
        COLOR_OBJETIVO
    )

    # -----------------------------------
    # CONFIGURACIÓN
    # -----------------------------------

    crear_caja(
        3.0, 9.0, 4.0, 0.9,
        "1. Configuración del Entorno",
        "Google Colab y Librerías",
        COLOR_CONFIG
    )

    # -----------------------------------
    # INGESTA E INSPECCIÓN
    # -----------------------------------

    crear_caja(
        3.0, 7.5, 4.0, 0.9,
        "2. Ingesta e Inspección",
        "read_excel() | head() | info()",
        COLOR_PROCESO
    )

    # -----------------------------------
    # LIMPIEZA Y TRANSFORMACIÓN
    # -----------------------------------

    crear_caja(
        3.0, 6.0, 4.0, 0.9,
        "3. Limpieza y Transformación",
        "scripts/limpieza.py",
        COLOR_PROCESO
    )

    # -----------------------------------
    # OPERACIONES
    # -----------------------------------

    crear_caja(
        0.8, 4.0, 3.8, 1.0,
        "4-A. Operaciones",
        "Análisis operacional",
        COLOR_OPERACIONES
    )

    # -----------------------------------
    # GESTIÓN AMBIENTAL
    # -----------------------------------

    crear_caja(
        5.4, 4.0, 3.8, 1.0,
        "4-B. Gestión Ambiental",
        "Cumplimiento DBO",
        COLOR_AMBIENTAL
    )

    # -----------------------------------
    # DASHBOARD
    # -----------------------------------

    crear_caja(
        3.0, 2.0, 4.0, 1.0,
        "5. Dashboard Exploratorio",
        "Plotly Express y Seaborn",
        COLOR_DASHBOARD
    )

    # -----------------------------------
    # GITHUB
    # -----------------------------------

    crear_caja(
        3.0, 0.5, 4.0, 0.9,
        "6. Control de Versiones",
        "Repositorio GitHub",
        COLOR_GITHUB
    )

    #========================
    # FLECHAS
    #========================

    # Objetivo → Configuración
    crear_flecha(5.0, 10.5, 5.0, 9.9)

    # Configuración → Ingesta
    crear_flecha(5.0, 9.0, 5.0, 8.4)

    # Ingesta → Limpieza
    crear_flecha(5.0, 7.5, 5.0, 6.9)

    # Limpieza → Operaciones
    crear_flecha(4.5, 6.0, 2.7, 5.0)

    # Limpieza → Ambiental
    crear_flecha(5.5, 6.0, 7.3, 5.0)

    # Operaciones → Dashboard
    crear_flecha(2.7, 4.0, 4.5, 3.0)

    # Ambiental → Dashboard
    crear_flecha(7.3, 4.0, 5.5, 3.0)

    # Dashboard → GitHub
    crear_flecha(5.0, 2.0, 5.0, 1.4)

    #========================
    # EXPORTAR IMÁGENES
    #========================

    plt.savefig(
        "flujo_trabajo_aqualimpia.png",
        dpi=300,
        bbox_inches='tight'
    )

    plt.savefig(
        "flujo_trabajo_aqualimpia.pdf",
        bbox_inches='tight'
    )

    #========================
    # MOSTRAR RESULTADO
    #========================

    plt.show()

    print("Flujo de trabajo generado correctamente.")
    print("Archivos exportados:")
    print("• flujo_trabajo_aqualimpia.png")
    print("• flujo_trabajo_aqualimpia.pdf")


#============================
# EJECUTAR FUNCIÓN
#============================

crear_flujo_trabajo()
```
